```python
# random_seeds = [random.randint(0, 10000) for i in range(10)]
seeds = [9779,8849,3869,1954,6662,3894,6302,2571,7234,1207]
```

```python
results = []
```

```python
k = 1
for seed in seeds:
    clientes       = Clientes(ncl=1000, propc=[0.25, 0.3, 0.45], propg=0.75, seed=seed)
    centrales      = Centrales(centrales_por_tipo=[5, 10, 25], seed=seed)
    parametros     = ProblemParameters(clients_vector=clientes, power_plants_vector=centrales)
    estado_inicial = InitialState.simple_state(parametros)
    problem        = EnergyProblem(estado_inicial, used_actions=[0], used_heuristic=COMBINED_HEURISTIC)

    start        = time.time()
    estado_final = hill_climbing(problem)
    end          = time.time()
    results.append([estado_inicial, end - start, estado_final, seed])

    print(f"Ejecucion {k}")
    k += 1
```

```python
for i in results:
        print("\n-------------------------------------------\n")
        print(f"Estado iniciail: {i[0]}")
        print(f"Tiempo de ejecución: {i[1]}\n")
        print(f"Estado final: {i[2]}")
        print(f"Uso de operadores: {i[2].count_actions}")
        print(f"Semilla: {i[3]}")
        print("\n-------------------------------------------\n")
```
